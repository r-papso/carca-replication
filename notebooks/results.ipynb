```python
import os
import matplotlib.pyplot as plt
```

```python
base_dir = "../results/video_games/"
dirs = os.listdir(base_dir)
```

```python
csv = next(os.path.join(base_dir, dirs[0], f) for f in os.listdir(os.path.join(base_dir, dirs[0])) if f.endswith(".csv"))
```

```python
data = dict()

for d in dirs:
    csv = next(os.path.join(base_dir, d, f) for f in os.listdir(os.path.join(base_dir, d)) if f.endswith(".csv"))
    table = []

    with open(csv, "r") as f:
        for l in f:
            vals = l.strip().split(";")
            table.append(vals)

    data[d] = table
```

```python
[round(i * 0.1, 1) for i in range(11)]
```

```python
plt.figure(figsize=(12, 10))

for k, v in data.items():
    plt.plot([int(t[1]) for t in v if t[2] == "val"], [float(t[4]) for t in v if t[2] == "val"], label=k)

plt.xticks([i * 10 for i in range(15)])
plt.yticks([round(i * 0.1, 1) for i in range(2, 9)])
plt.legend()
plt.show()
```
